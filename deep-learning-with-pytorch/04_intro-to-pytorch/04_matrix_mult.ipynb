```python
import torch
```

```python
def activation(x):
    """ Sigmoid activation function
    
        Arguments
        ---------
        x: torch.Tensor
    """
    return 1/(1 + torch.exp(-x))
```

### Generate some data

```python
torch.manual_seed(7)  # Set the random seed so things are predictable

# Features are 5 random normal variable
features = torch.randn((1, 5))
# True weights for our data, random normal variable again
weights = torch.randn_like(features)
# and a true bias term
bias = torch.randn((1, 1))
```

### Solution -- Element-wise

```python
y = activation(torch.sum(features * weights) + bias)
print(y)

y = activation((features * weights).sum() + bias)
print(y)
```

### Solution -- Using Matrix Multiplication

```python
# Checking shape
w = weights.view(5, 1)  # reshaped

y = activation(torch.mm(features, w) + bias)
print(y)
```
